```python
X = df.drop(['Class'], axis=1) # X represents all the features
y = df['Class'] # y represents the target variable (class).
```

```python
# scaling all the features
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
# plotting all the features again after scaling.
fig, ax = plt.subplots(ncols=5, nrows=6, figsize=(30, 30)) 
ax = ax.flatten()

for index in range(30):
    sns.distplot(X_scaled[:, index], ax=ax[index])
plt.tight_layout(pad=0.5, w_pad=0.5, h_pad=0.5)
```

```python
# splitting the dataset for training and testing.
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=0.8) # keeping 80% training data and rest as testing data.
```

```python
# importing different evaluation metrics.
from sklearn.metrics import classification_report, roc_auc_score, f1_score, precision_score, accuracy_score, recall_score
```

```python
# using logistic regression to predict the output.
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(n_jobs=-1)
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
```

```python
print('For Logistic Regression:')
print('ROC AUC score =', roc_auc_score(y_test, y_pred))
print('F1 score =', f1_score(y_test, y_pred))
print('Precision score =', precision_score(y_test, y_pred))
print('Recall score =', recall_score(y_test, y_pred))
print('Accuracy score =', accuracy_score(y_test, y_pred))
```

```python
# using Bagging Classifier to predict the output.
from sklearn.ensemble import BaggingClassifier

bc = BaggingClassifier(n_jobs=-1)
bc.fit(x_train, y_train)
y_pred = bc.predict(x_test)
```

```python
print('For Bagging Classifier:')
print('ROC AUC score =', roc_auc_score(y_test, y_pred))
print('F1 score =', f1_score(y_test, y_pred))
print('Precision score =', precision_score(y_test, y_pred))
print('Recall score =', recall_score(y_test, y_pred))
print('Accuracy score =', accuracy_score(y_test, y_pred))
```

```python
# using Random Forest Classifier to predict the output.
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_jobs=-1)
rfc.fit(x_train, y_train)
y_pred = rfc.predict(x_test)
```

```python
print('For Random Forest Classifier:')
print('ROC AUC score =', roc_auc_score(y_test, y_pred))
print('F1 score =', f1_score(y_test, y_pred))
print('Precision score =', precision_score(y_test, y_pred))
print('Recall score =', recall_score(y_test, y_pred))
print('Accuracy score =', accuracy_score(y_test, y_pred))
```

```python
# using XGB Classifier to predict the output.
from xgboost import XGBClassifier

xgbc = XGBClassifier(objective='binary:logistic')
xgbc.fit(x_train, y_train)
y_pred = xgbc.predict(x_test)
```

```python
print('For XGB Classifier:')
print('ROC AUC score =', roc_auc_score(y_test, y_pred))
print('F1 score =', f1_score(y_test, y_pred))
print('Precision score =', precision_score(y_test, y_pred))
print('Recall score =', recall_score(y_test, y_pred))
print('Accuracy score =', accuracy_score(y_test, y_pred))
```

```python
# using over-sampling to treat imbalanced dataset.
from imblearn.over_sampling import SMOTE

over_sample = SMOTE()
x_smote, y_smote = over_sample.fit_resample(x_train, y_train) # x_smote and y_smote is now new training data.
```

```python
# plotting the occurences of each class after over-sampling.
sns.countplot(x=y_smote)
plt.show()
```

```python
# using Random Forest Classifier to predict the output.
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_jobs=-1)
rfc.fit(x_smote, y_smote)
y_pred = rfc.predict(x_test)
```

```python
print('For Random Forest Classifier:')
print('ROC AUC score =', roc_auc_score(y_test, y_pred))
print('F1 score =', f1_score(y_test, y_pred))
print('Precision score =', precision_score(y_test, y_pred))
print('Recall score =', recall_score(y_test, y_pred))
print('Accuracy score =', accuracy_score(y_test, y_pred))
```

```python
# using XGB Classifier to predict the output.
from xgboost import XGBClassifier

xgbc = XGBClassifier(objective='binary:logistic')
xgbc.fit(x_smote, y_smote)
y_pred = xgbc.predict(x_test)
```

```python
print('For XGB Classifier:')
print('ROC AUC score =', roc_auc_score(y_test, y_pred))
print('F1 score =', f1_score(y_test, y_pred))
print('Precision score =', precision_score(y_test, y_pred))
print('Recall score =', recall_score(y_test, y_pred))
print('Accuracy score =', accuracy_score(y_test, y_pred))
```
